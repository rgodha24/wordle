```python
import json
from random import randint 
from tqdm import tqdm_notebook as tqdm
import time

with open("answers.json") as answerJson:
    answers = json.load(answerJson)
    
with open("ok.json") as okJson:
    ok = json.load(okJson)
    
    
greens = []
greens.append(input("first letter if it is green").strip().lower())
greens.append(input("second letter if it is green").strip().lower())
greens.append(input("third letter if it is green").strip().lower())
greens.append(input("fourth letter if it is green").strip().lower())
greens.append(input("fifth letter if it is green").strip().lower())

for i in range(len(greens)):
    if len(greens[i]) == 0:
        greens[i] = None

yellows = input("all yellow letters seperated by a space").strip().lower().split()

greys = input("all grey letters seperated by space").lower().strip().split()

cantBe = []
for i in range(5):
    cantBe.append(input(f"what letters can number {i+1} not be seperated by space").strip().lower().split())
    
for i in range(len(cantBe)):
    if len(cantBe[i]) == 0:
        cantBe[i] = None

print(greens, yellows, greys)
```

```python

# checking if greens are ok
possibleAnswers = []
for answerInt, answerX in enumerate(answers):
    fine = True
    for letterInt, letter in enumerate(answerX):
        if greens[letterInt] != None:
            if greens[letterInt] != letter:
                fine = False
                
    if fine:
        possibleAnswers.append(answerX)
        
# print(len(possibleAnswers))
# checking yellows

if len(yellows) != 0:
    popList = []
    for answerInt, answerX in enumerate(possibleAnswers):
        a = []
        for letterInt, letterX in enumerate(answerX):
            if letterX in yellows:
                a.append(True)
        
        if len(a) != len(yellows):
            popList.append(answerInt)
            
    for bad in reversed(popList):
        try:
            possibleAnswers.pop(bad)
        except:
            print(f"pop failed LLLLLLL {bad}")
        
if len(greys) != 0:
    popList = [] #indexes of bad ones
    for answerInt, answerX in enumerate(possibleAnswers):
        for letterInt, letterX in enumerate(answerX):
            if letterX in greys:
                popList.append(answerInt)
                break
                
    for bad in reversed(popList):
        try:
            possibleAnswers.pop(bad)
        except:
            print(f"pop failed LLLLLLL {bad}")

    # print(len(possibleAnswers))
    
if len(cantBe) != 0:
    popList = []
    for answerInt, answerX in enumerate(possibleAnswers):
        for letterInt, letterX in enumerate(answerX):
            if len(cantBe[letterInt]) != 0:
                if letterX in cantBe[letterInt]:
                    popList.append(answerInt)
                    
    for bad in reversed(popList):
        try:
            possibleAnswers.pop(bad)
        except:
            print(f"pop failed LLLLLLL {bad}")
    
    # print(len(possibleAnswers))
    
    

if len(possibleAnswers) > 20:
    print(f"possible answers is {len(possibleAnswers)}. not gonna write that many")
elif len(possibleAnswers) > 1000:
    print("first word is going to be: ")
else:
    print(possibleAnswers)
```

```python
#finding the best word to put in
# ranking the letters (copied from best starting word)

def betterIndex(list, finding):
    outputList = []
    for i, j in enumerate(list):
        if j == finding:
            outputList.append(i)
            
    return outputList

def rank(ausdhb):
    list = ausdhb.copy()
    x = []
    y = list.copy()
    popped = []
    for i in range(len(list)):
        a = max(list)
        b = list.index(a)
        
        
        appending = None
        for j, k in enumerate(betterIndex(y, a)):
            if k not in popped:
                appending = k
        x.append(appending)
        
        
        popped.append(appending)
        list.pop(b)
    # print(popped)    
    # print(list)
    return x

letterList = []
x = []
for q in range(26):
    x.append(0)
    
for w in range(5):
    letterList.append(x.copy())

for j, k in enumerate(possibleAnswers):
    for letterNumber, letter in enumerate(k):
        letterList[letterNumber][ord(letter)-97] += 1
        
# print(letterList)

ranks = []
for i in range(5):
    ranks.append(rank(letterList[i]))
    
letterRanks = []
for i in ranks:
    o = []
    for j in i:
        try:
            o.append(chr(j+97))
        except:
            print("one of them failed lol awef jnaow pain")
        
    letterRanks.append(o)

# letter ranks is now a rank of the best letter for each part of the word. 
# now we have to generate the best word to put in next with this list.
```
